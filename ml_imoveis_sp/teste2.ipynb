```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from dotenv import load_dotenv

load_dotenv()

px.set_mapbox_access_token(open('mapbox_token').read())
```

```python
df_data = pd.read_csv("sao-paulo-properties-april-2019.csv")
df_rent = df_data[df_data["Negotiation Type"] == "rent"]
df_cleaned = df_rent.drop(["New", "Property Type", "Negotiation Type"], axis=1)
```

```python
# Tratamento de Dados Categóricos
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

# aqui essa função vai pegar as strings e aplicar um numero que a representa e a armazena
# para criar um DataFrame podemos passar uma lista de um unico valor para dentro de colchetes
district_encoded = ordinal_encoder.fit_transform(df_rent[["District"]])
district_encoded
```

```python
# Tratamento de Dados Categóricos
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()


# aqui atribuimos cada distrito/bairro que seja diferente ele vira uma coluna e se o bairro corresponder ao ap aplicaremos o 1 para sim e 0 para não (boleano)
housing_cat_hot = cat_encoder.fit_transform(df_cleaned[["District"]])
housing_cat_hot

# sparse matrix = matriz muito grande com muitos 0, essa questão deleta onde tem 0 e so guarda onde tem 1
```

```python
# tratando a função de sparse matrix com pandas
one_hot = pd.get_dummies(df_cleaned["District"])

df = df_cleaned.drop('District', axis=1)
df = df.join(one_hot)
df
```
